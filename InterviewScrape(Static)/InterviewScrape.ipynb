```python
def interviewScrape(soup):
    name = [i.find_all('h5', class_ = 'card-title')[0].text.strip().split(',')[0] for i in soup.select('div.card-body')]
    position = [i.find_all('h5', class_ = 'card-title')[0].text.strip().split(',')[1].split('at')[0].strip() for i in soup.select('div.card-body')]
    cmp = []
    company = [i.find_all('h5', class_ = 'card-title')[0].text.strip().split(',')[1] for i in soup.select('div.card-body')]
    for r in company:
        s = r.find('at')
        cmp.append(r[s+2:])
    quote = [i.find_all('p', class_='card-text')[0].text.strip() for i in soup.select('div.card-body')]
    dates = []
    for r in soup.select('div.card-body'):
        dates.append(r.find_all('p', class_='card-text')[1].text.split()[:3])
        date = []
        for d in dates:
            date.append("".join(d))
    dates = pd.to_datetime(date, format = '%B%d,%Y')
    tags = []
    for r in soup.select('div.card-body'):
        q = r.find_all('strong', class_ ='mr-2')
        tags.append([i.text.strip() for i in q])
    return pd.DataFrame({'Name':name, 'Job Position': position,  'Company':cmp,
                        'Quote': quote, 'Date': dates, 'Tags': tags})
```

```python
interviewScrape(soup)
```

```python
def interviewScraping(n):
    info = []
    page = 'https://devtomanager.com/interviews/page/{}/'
    for i in range(1,n+1):
        html = requests.get(page.format(i)).text
        time.sleep(3)
        soup = BeautifulSoup(html, 'html.parser')
        info.append(interviewScrape(soup))
    df = pd.concat(info, axis = 0, ignore_index = True)
    return df.to_csv('InterviewScrap.csv')
```

```python
interviewScraping(5)
```
